```python
import scipy.interpolate as scint
```

```python
xgrid,ygrid=np.mgrid[0:2.001:1,0:2.001:1]
# Input grid
xgrid1,ygrid1=np.mgrid[0:2.001:0.01,0:2.001:0.01]
#output grid
```

```python
Func = lambda x,y: x**2+(y-0.2)**3+0.5*x*y
# function to interpolate
```

```python
zgrid = Func(xgrid,ygrid)
```

```python
RegI = scint.RegularGridInterpolator((xgrid[:,0].flatten(),ygrid[0,:].flatten()),zgrid)
# Linear (or multilinear interpolation)
```

```python
TriI = scint.LinearNDInterpolator(np.array([xgrid.flatten(),ygrid.flatten()]).T,zgrid.flatten())
# Actual linear interpolation using Delaunay triangulation
```

```python
TriV = TriI(xgrid1,ygrid1)
RegV = RegI((xgrid1,ygrid1))
```

```python
plt.imshow(TriV);
```

```python
plt.imshow(RegV);
```

```python
plt.imshow(TriV-RegV);
plt.colorbar();
plt.title('Difference of interpolations')
```
